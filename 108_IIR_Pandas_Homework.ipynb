```python
import pandas as pd
```

```python
iris = pd.read_csv('./iris.csv')
iris.columns = ['A', 'B', 'C', 'D', 'class']
```

```python
def text_function(row):
    return 'class: {class}, A:{A},B:{B},C:{C},D:{D}'.format(**row.to_dict())
```

```python
%%timeit 
print(iris.apply(text_function, axis=1))
```

```python
%%timeit 
for ir in iris.itertuples():
    print(ir)
```
